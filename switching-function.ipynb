```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def switch_func(r,r_s=0.5,r_c=6.0):
    s=np.zeros_like(r)
    x=np.zeros_like(r)
    mask1=(r<=r_s)
    s[mask1]=1/r[mask1]
    mask2=(r_s<=r) & (r<r_c)
    x[mask2]=(r[mask2]-r_s)/(r_c-r_s)
    s[mask2]=(0.5*np.cos(np.pi*x[mask2])+0.5)/r[mask2]
    return s

def switch_func_new(r,r_s=0.5,r_c=6.0):
    s=np.zeros_like(r)
    x=np.zeros_like(r)
    mask1=(r<=r_s)
    s[mask1]=1/r[mask1]
    mask2=(r_s<=r) & (r<r_c)
    x[mask2]=(r[mask2]-r_s)/(r_c-r_s)
    s[mask2]=(x[mask2]**3*(-6*x[mask2]**2+15*x[mask2]-10)+1)/r[mask2]
    return s
```

```python
r_s,r_c=0.5,6.0
r=np.linspace(0.1,7.0,50)
plt.plot(r,switch_func(r,r_s=0.5,r_c=6.0))
# plt.plot(r,switch_func_new(r,r_s=0.5,r_c=6.0))
plt.axvline(x=r_s,color="red",linestyle="--")
plt.axvline(x=r_c,color="green",linestyle="--")
# plt.ylim(0,0.5)
plt.show()
```
